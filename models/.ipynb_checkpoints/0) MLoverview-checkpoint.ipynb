### Reinforement Learning

> In this setting, an agent learns to achieve a desired state/goal using smart trial and error by taking various actions and a reward/regret/punishment is provided as it takes these actions.

> The rewards/punishments might be sparse, i.e they might not be available at each step the agent takes an action. The rewards might only be available after a long sequence of steps.

>These kind of models are mostly used in robotics as labels cannot be provided for each step an action is taken by the robot. For example, we cannot teach a robot step by step how to walk. We can let it learn by itself if we provide a reward to it if it succeeds to walk, say 10 meters without falling. In this way, the robot can learn what sequence of steps are beneficial in learning to walk after a lot of smart trail and error moves.

<h4 style="color:cyan"> We will be focusing on Supervised ML, the most common one</h4>

> Supervised learning can further be classified into two categories
> 1) Regression
> 2) Classification

> **Regression:** concerns with predicting continuous real valued signals from given features/information. For example, one might want to predict the rental price of a house given some information about the city, the location in the city, the floor area size, the number of bedrooms, number of washrooms, garage/parking space, other amenities, etc. Since the rental price of a house is a continous real value, it is a regression problem.
